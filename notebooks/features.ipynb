```python
def count_tokens(text_str):
    """This function takes strings and return the number of tokens

    Args:
        text_str (str): string of texts 
    """
    # initialize tokernizer
    tokenizer = GPT2TokenizerFast.from_pretrained("gpt2")
    num_tokens = len(tokenizer(text_str)['input_ids'])
    return(num_tokens)

count_tokens("tessacasca")
```

```python
def calculate_total(d):
    total = 0
    for value in d.values():
        if isinstance(value, dict):
            total += calculate_total(value)
        elif isinstance(value, list):
            total += sum(value)
    return total

data = {
    "level1a": {
        "level2a": {
            "level3a": {
                "level4a": [1, 2, 3, 4],
            },
            "level3b": {
                "level4b": [5, 6, 7, 8],
            },
        },
        "level2b": {
            "level3c": {
                "level4c": [9, 10, 11, 12],
            },
        },
    },
    "level1b": {
        "level2c": {
            "level3d": {
                "level4d": [13, 14, 15, 16],
            },
            "level3e": {
                "level4e": [17, 18, 19, 20],
            },
        },
    },
}

total = calculate_total(data)
print(total)
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sumy.summarizers.lex_rank import LexRankSummarizer
from sumy.summarizers.text_rank import TextRankSummarizer
from sumy.summarizers.luhn import LuhnSummarizer
from sumy.summarizers.lsa import LsaSummarizer
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer

def count_sentences(text):
    return len([sentence for sentence in text.split('. ') + text.split('! ') + text.split('? ') if sentence.strip()])

def summarize_text(text):
    num_sent = count_sentences(text)
    print(num_sent)
    # stop_words = set(stopwords.words('english'))
    # words = word_tokenize(text)
    # words = [word.lower() for word in words if word.isalpha()]
    # words = ' '.join([word for word in words if word not in stop_words])
    parser = PlaintextParser.from_string(text,Tokenizer("english"))
    summarizer = TextRankSummarizer()
    summarized_text = summarizer(parser.document, num_sent//3)
    return " ".join([str(sentence) for sentence in summarized_text])

text = "The survey was conducted from February 2021 to April 2022, during which some 2694 assessment reports were collected from 165 participants, for 65 survey points in four typical residential areas (see Section 2.2.2), 90% of the questionnaires were completed between January and June 2021, and the addition of the new content did not change the overall questionnaire much. 50.30% of the participants were males and 49.70% females, 87.27% had higher education, and 53.33% were aged 18 to 30 years (Table 3 ). As this study aims to reveal the relationship between FoS/PLQ and light attributes, only assessments of these variables were used in the following analysis."
summarize_text(text)
```
